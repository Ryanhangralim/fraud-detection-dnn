### Import Libraries

### Read Files (Gathering Data)

### Data Cleaning

```python
# Remove columns with at least 00% null values
transaction = transaction.dropna(axis=1, thresh=len(transaction) * 0.9)

# Display remaining columns
print(transaction.columns)
```

```python
transaction.isnull().sum()
```

```python
transaction.isFraud.value_counts()
```

```python
# Identify columns with some null values
columns_with_nulls = transaction.columns[transaction.isnull().any()]

# Fill null values with 0 in those columns
transaction[columns_with_nulls] = transaction[columns_with_nulls].fillna(0)

# Verify that there are no null values in the dataset
print(transaction.isnull().sum())
```

```python
transaction.isnull().sum()
```

```python
transaction = transaction.drop("TransactionID", axis=1)
```
