### Import Libraries

```python
import numpy as np 
import pandas as pd
```

### Read Files (Gathering Data)

```python
transaction = pd.read_csv("dataset/train_transaction.csv")
```

### Assessing Data

### Data Cleaning

```python
# Remove columns with at least 00% null values
removednull_transaction = transaction.dropna(axis=1, thresh=len(transaction) * 0.9)

# Display remaining columns
print(removednull_transaction.columns)
```

```python
removednull_transaction.isnull().sum()
```

```python
removednull_transaction.isFraud.value_counts()
```

```python
# Identify columns with some null values
columns_with_nulls = removednull_transaction.columns[removednull_transaction.isnull().any()]

# Fill null values with 0 in those columns
removednull_transaction[columns_with_nulls] = removednull_transaction[columns_with_nulls].fillna(0)

# Verify that there are no null values in the dataset
print(removednull_transaction.isnull().sum())
```

```python
removednull_transaction.isnull().sum()
```
